```python
import pandas as pd
import numpy as np

df=pd.read_csv("emp.csv")

df[["salary","designation"]]
df[df["salary"] > 50000]
df.query("department == 'IT' and salary > 50000")[["name", "salary"]]
df["salary"].max()
df[df["salary"]==df["salary"].max()]#disp details of emp with max salary
```

```python
df.groupby("city").size()
```

```python
df[df["salary"].isna()] #salary is null
df["salary"].mean()
df["salary"].fillna(df["salary"].mean(), inplace=True)
```

```python
import matplotlib.pyplot as plt
plt.bar(df['experience'],df['salary'])
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title("Experience vs Salary")
plt.show()
```

```python
import matplotlib.pyplot as plt

age_counts = df["age"].value_counts().sort_index()

plt.figure()
plt.bar(age_counts.index, age_counts.values)

plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Employee Count by Age")

plt.show()
```
